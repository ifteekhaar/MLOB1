# Exceptions


1. ### Handling Exceptions in python

exceptions are used to handle errors SO that the program dont crash


To solve error we use try and except block

```python
numbers = [1,2]
print(numbers[3])

# This will show an index error as there are only index 0 and 1.
# no index 3
# to handle this error we need to make an exception
```

```python
# Fix:
numbers = [1,2]
try: 
    print(numbers[3])
except IndexError:
    print("The index is not there")
    
# Now the program wont crash. 
# We can add multiple exception
```

```python
age = int(input("Enter age : "))

# Here if we enter a non integer or a character 
# it will show valueError 
```

```python
try: 
    age = int(input("Enter age : "))
    
except ValueError: 
    print("Wrong input")
    
except IndexError:
    print("Index not available")
else:
    print("No error.execution continues")
```

```python
numbers = [1,2]
try: 
    print(numbers[3])
except IndexError:
    print("Error Occured")
    print(IndexError)
    print(type(IndexError))
```

```python
try:
    value = int(input("Enter value: "))
    division = 30/ value
except (ValueError, ZeroDivisionError):
    print("Error Occured")
else:
    print("No error occured")
    
```

2. ### Cleaning up

There are times we need to work with external resourses like database. 

Whenever we are done with this resourses we need to release or close them. 

Like after opening a file we must close it 



```python
try:
    file = open("5_exceptions.ipynb")
    
    value = int(input("Enter value: "))
    division = 30/ value
    
except (ValueError, ZeroDivisionError):
    print("Error Occured")
    
else:
    print("No error occured")
    
finally:
    file.close()
```

4. ### Raising Esceptions

```python
def calculate_xfactor(age):
    if age <= 0 :
        raise ValueError("Age cant be zero or negative")  # raising exception
    return 10 / age

try: 
    calculate_xfactor(-10)
except ValueError as error:
    print(error)
```

5. ### Cost of raising exceptions


raising exception is not preferable as it efects the preformance in large scale model 


```python
from timeit import timeit

# This function can calculate the execution time 
# lets calculate the execution time of the previous one
# to do so lets define a variable 
# that contains the python code in a string
```

```python
from timeit import timeit

code1 = '''

def calculate_xfactor(age):
    if age <= 0 :
        raise ValueError("Age cant be zero or negative")
    return 10 / age

try: 
    calculate_xfactor(-10)
except ValueError as error:
    print(error)
    
'''

print("first_code=", timeit(code1, number= 10000))    # running the code 1000 time
```

```python
# Lets see without raising exception

from timeit import timeit

code1 = '''

def calculate_xfactor(age):
    if age <= 0 :
        return none
    return 10 / age


xfactor = calculate_xfactor(-10)
if xfactor == none:
    pass
    
'''

print("first_code=", timeit(code1, number= 10000))
```
